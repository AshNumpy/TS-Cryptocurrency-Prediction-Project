```python
import pandas as pd 
import numpy as np 
import yfinance as yf
from datetime import date, timedelta
today = date.today()
end_date = today.strftime('%Y-%m-%d')

d2 = date.today() - timedelta(days=365)
start_data = d2.strftime('%Y-%m-%d')

data = yf.download('BTC-USD', start=start_data, end=end_date, progress=False)
data.reset_index(inplace=True)
data.tail()
```

```python
data.shape
```

```python
import plotly.graph_objects as go 
figure = go.Figure(data=[go.Candlestick(
    x=data['Date'],
    open=data.Open,
    high=data.High,
    low=data.Low,
    close=data.Close
)])
figure.update_layout(
    title='BTC Price Analysis',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)
figure.show()
```

```python
correlation = data.corr()
correlation.Close.sort_values(ascending=False)
```

```python
### MODEL ###
from autots import AutoTS
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
